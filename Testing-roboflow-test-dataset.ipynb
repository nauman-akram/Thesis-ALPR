```python
import requests
import os
import time
import pandas as pd
from pprint import pprint
```

```python
os.getcwd()
```

```python
def call_PRAPI(file_path):
    
    with open(file_path, 'rb') as fp:

        response = requests.post(
            'https://api.platerecognizer.com/v1/plate-reader/',
            #data=dict(regions=regions),  # Optional
            files=dict(upload=fp),
            headers={'Authorization': 'Token 5ea9aaf15a07f768e4959ece52955592138fb925'})
        result = response.json()
    return result
```

```python
dataset_path = os.path.join(os.getcwd(),'dataset\\test')
dataset_path
```

```python
df = pd.read_csv(dataset_path+'\\_annotations.csv')
```

```python
df
```

```python
detection = []
for fn in df.filename:
    img_path = dataset_path+'\\'+fn
    #print(img_path)
    #show_image(dataset_path+'\\'+fn)
    response = call_PRAPI(img_path)
    print(response)
    if response['results']:
        lp =  response['results'][0]['plate'] 
        print(lp)
        detection.append( lp)
    else:
        detection.append('not found')

    time.sleep(1.1)
```

```python
df['PR_det'] = detection
```
